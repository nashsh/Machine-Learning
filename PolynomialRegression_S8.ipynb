```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
ds=pd.read_csv('Position_Salaries.csv')
X=ds.iloc[:,1:-1].values
y=ds.iloc[:,-1]
```

```python
from sklearn.linear_model import LinearRegression
lin_regressor = LinearRegression()
lin_regressor.fit(X,y)
```

Training the Polynomial Regression model on the whole dataset

```python
from sklearn.preprocessing import PolynomialFeatures
poly_regressor= PolynomialFeatures(degree=2)
X_poly = poly_regressor.fit_transform(X)
X_poly
lin_reg_2 = LinearRegression()
lin_reg_2.fit(X_poly,y)
```

Visualising the Linear Regression Result

```python
plt.scatter(X,y,color='red')#real result
plt.plot(X,lin_regressor.predict(X),color='blue')#predictions
plt.title('Truth or Bluff (linear regresssion model)')
plt.xlabel('position level')
plt.ylabel('salary')
plt.show()
```

```python
plt.scatter(X,y,color='red')
plt.plot(X,lin_reg_2.predict(X_poly),color='blue')
plt.xlabel('position level')
plt.ylabel('salary')
plt.show()
```

```python
from sklearn.preprocessing import PolynomialFeatures
poly_reg_higher=PolynomialFeatures(degree=4)
X_poly_higher=poly_reg_higher.fit_transform(X)
lin_reg_higher=LinearRegression()
lin_reg_higher.fit(X_poly_higher,y)
```

```python
plt.scatter(X,y,color='red')
plt.plot(X,lin_reg_higher.predict(X_poly_higher),color='blue')
```

Smoother curve and higher resolution visualising

Prediciting a new result in Linear Regression

```python
print(lin_regressor.predict([[6.5]]))
```

Prediciting a new result in Polynomial Regression

```python
print(lin_reg_2.predict(poly_regressor.fit_transform([[6.5]])))
```
