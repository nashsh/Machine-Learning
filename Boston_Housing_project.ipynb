Importing libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

Loading the Dataset

```python
url='https://raw.githubusercontent.com/dataprofessor/data/refs/heads/master/BostonHousing.csv'
dataset = pd.read_csv(url)
```

```python
print(dataset.columns)
```

```python
X= dataset.iloc[:, :-1].values
y= dataset.iloc[:, -1].values
```

Splitting Dataset into training and test set

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test= train_test_split(X,y,test_size=0.2,random_state=0)
```

Linear Regression Model

```python
from sklearn.linear_model import LinearRegression
regressor = LinearRegression()
regressor.fit(X_train,y_train)
```

```python
y_predict=regressor.predict(X_test)
np.set_printoptions(precision=2)
print(np.concatenate((y_predict.reshape(len(y_predict),1),y_test.reshape(len(y_test),1)),axis=1))
```

```python
import seaborn as sns
sns.scatterplot(x=y_test,y=y_predict,alpha=0.5)
sns.scatterplot(x=[0], y=[regressor.intercept_], color='red')
```

```python


# Assuming Y_test and Y_predict are the actual and predicted values
Y_testt = np.array(y_test)  # Replace with your actual Y values

# Plot the actual vs. predicted values
sns.scatterplot(x=Y_testt, y=y_predict, color='blue', label='Predicted vs. Actual',alpha=0.5)

# Plot a 45-degree line for reference
max_val = max(Y_testt.max(), y_predict.max())
min_val = min(Y_testt.min(), y_predict.min())
plt.plot([min_val, max_val], [min_val, max_val], 'r--', label='Ideal fit line')

# Labels and legend
plt.xlabel('Actual Values')
plt.ylabel('Predicted Values')
plt.legend()
plt.show()
```
